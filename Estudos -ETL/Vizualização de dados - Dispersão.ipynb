```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
#criando gráfico de dispersão considerando volume e circunferência
plt.scatter(x= base.Girth, y= base.Volume)
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
#usando o seaborn
sbn.regplot(x= base.Girth, y=base.Volume, data= base, x_jitter= 0.1, fit_reg= True)
```
