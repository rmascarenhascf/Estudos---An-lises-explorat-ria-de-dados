```python
import pandas as pd
```

```python
base= pd.read_csv('insect.csv')
display(base.head())
base.shape
```

```python
base.groupby(['spray']).size()
```

```python
# Agrupar dados baseado no atributo 'spray', contando e somando os registros
agrupado = base.groupby(['spray'])['count'].sum()
agrupado
```

```python
# gráfico de barras
agrupado.plot.bar(color= 'gray')
```

```python
# cores
agrupado.plot.bar(color= ['blue', 'yellow','red','green', 'purple','orange'])
```
