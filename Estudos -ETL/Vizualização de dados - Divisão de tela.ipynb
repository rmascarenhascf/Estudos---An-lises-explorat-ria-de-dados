```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
#girth com volume
plt.scatter(base.Girth, base.Volume)
```

```python
#Girth com Height
plt.scatter(base.Girth, base.Height, marker= '*')
```

```python
#histograma de volume
plt.hist(base.Volume)
```

```python
#imprimindo juntos
#Criao de figura, no qual os grficos sero posicionados

plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height, marker= '*')
plt.subplot(2,2,3)
plt.hist(base.Volume)
```
