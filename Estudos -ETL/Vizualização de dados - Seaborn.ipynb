```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
#criando histograma com 10 divisões com matplotlib
plt.hist(base.iloc[:,1], bins= 10)
```

```python
# criando o histograma com linha de distribuição de frequência, com 6 divisões (bins)
#kde = linha de densidade

sbn.histplot(base.iloc[:,1], kde= False, bins= 6, color= 'blue').set(title= 'Árvores')
```

```python
#criando com apenas a linha de densidade
sbn.kdeplot(base.iloc[:,1], color= 'blue').set(title= 'Árvore')
```

```python
#criando densidade e histograma
sbn.histplot(base.iloc[:,1], kde= True,bins= 6 , color= 'blue').set(title= 'Árvores')
```
